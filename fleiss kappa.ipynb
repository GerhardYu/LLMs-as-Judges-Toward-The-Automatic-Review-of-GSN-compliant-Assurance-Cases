```python
import numpy as np
import pandas as pd
from statsmodels.stats.inter_rater import fleiss_kappa
from statsmodels.stats.inter_rater import aggregate_raters
import openpyxl
```

```python
file = pd.read_excel('Classification of the results.xlsx', sheet_name='LLM Review Score')
file
```

```python
argument_comprehension = file.loc[file['Review Criterion'] == 'Argument Comprehension']
ZS_AC = argument_comprehension.loc[argument_comprehension['Prompting Strategies'] == 'Zero Shot']

ZS_AC_4o = ZS_AC.loc[ZS_AC['LLM'] == 'GPT4o']
ZS_AC_4o = ZS_AC_4o.drop(columns=['Review Criterion', 'Run', 'LLM', 'Prompting Strategies', 'Name of the System'])
ZS_AC_4o = ZS_AC_4o.to_numpy().reshape(20, 1)
#print(ZS_AC_4o)
#print(ZS_AC_4o.shape)

ZS_AC_4point1 = ZS_AC.loc[ZS_AC['LLM'] == 'GPT4.1']
ZS_AC_4point1 = ZS_AC_4point1.drop(columns=['Review Criterion', 'Run', 'LLM', 'Prompting Strategies', 'Name of the System'])
ZS_AC_4point1 = ZS_AC_4point1.to_numpy().reshape(20,1)
#print(ZS_AC_4point1)
#print(ZS_AC_4point1.shape)

ZS_AC_Deepseek = ZS_AC.loc[ZS_AC['LLM'] == 'Deepseek Reasoner']
ZS_AC_Deepseek = ZS_AC_Deepseek.drop(columns=['Review Criterion', 'Run', 'LLM', 'Prompting Strategies', 'Name of the System'])
ZS_AC_Deepseek = ZS_AC_Deepseek.to_numpy().reshape(20,1)
#print(ZS_AC_Deepseek)
#print(ZS_AC_Deepseek.shape)

ZS_AC_Gemini = ZS_AC.loc[ZS_AC['LLM'] == 'Gemini 2.0 Flash']
ZS_AC_Gemini = ZS_AC_Gemini.drop(columns=['Review Criterion', 'Run', 'LLM', 'Prompting Strategies', 'Name of the System'])
ZS_AC_Gemini = ZS_AC_Gemini.to_numpy().reshape(20,1)
#print(ZS_AC_Gemini)
#print(ZS_AC_Gemini.shape)

ZS_AC_aggregate = np.concatenate((ZS_AC_4o, ZS_AC_4point1, ZS_AC_Deepseek, ZS_AC_Gemini), axis = 1)
ZS_AC_aggregate = ZS_AC_aggregate.astype(int)
ZS_AC_aggregate = ZS_AC_aggregate - 1
#Aggregates are subtracted by 1 since aggregate_raters start from 0 and our ratings start at 1

aggregated_ZS_AC, categories= aggregate_raters(ZS_AC_aggregate, n_cat=5)
print(aggregated_ZS_AC)

fleiss_kappa_ZS_AC = fleiss_kappa(aggregated_ZS_AC, method="fleiss")
print("Fleiss Kappa: " ,fleiss_kappa_ZS_AC)
```
